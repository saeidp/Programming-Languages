# Manipulating Array Shapes

```python
import numpy as np
x=np.arange(9).reshape(3,3)
print(x)
print('----------')
ravelled_array = x.ravel()
print(ravelled_array)
print('----------')
flattened_array = x.flatten()
print(flattened_array)
# flatten always return a copy but ravel return a view of the original array
# if you change the ravelled in above the actual array gets changed as well
print('----------')

y = np.arange(9)
y.shape = [3,3]
print(y)
print('----------')

print(y.transpose())
print('----------')
print(y.T) # rthe same effect
print('----------')

print(np.resize(y, (6,6))) # data gets repeated
print('----------')

print(np.zeros((6,))) # one dimentional
print('----------')

print(np.zeros((6,), dtype = int)) # one dimentional
print('----------')

print(np.zeros((3,2), dtype = int)) 
print('----------')

print(np.ones((3,2), dtype = int)) 
print('----------')

print(np.eye(3)) 
print('----------')

print(np.random.rand(4,4)) # uniform distribution from zero and one
print('----------')
```

# Matrix Multiplication

```python
import numpy as np
mat_a= np.matrix([0,3,5,5,5,2]).reshape(2,3)
print(mat_a)
print('----------')
import numpy as np
mat_b= np.matrix([3,4,3,-2,4,-2]).reshape(3,2)
print(mat_b)
print('----------')

print(mat_a * mat_b)
print('----------')

product = np.matmul(mat_a, mat_b)
print(product)
print('----------')

#you can use the numpy array as well
mat_a= np.array([0,3,5,5,5,2]).reshape(2,3)
mat_b= np.array([3,4,3,-2,4,-2]).reshape(3,2)
product = np.matmul(mat_a, mat_b)
print(product)
print('----------')

print(mat_a @ mat_b)
print('----------')
```

# Stacking

```python
import numpy as np
x=np.arange(4).reshape(2,2)
print(x)
print('----------')

y=np.arange(4, 8).reshape(2,2) # fom 4 to 7
print(y)
print('----------')

z= np.hstack((x,y))
print(z)
print('----------')

x=np.arange(4).reshape(2,2)
print(x)
y = np.arange(8).reshape(4,2)
print(y)
z = np.vstack((x,y))
print(z)
print('----------')

w= np.concatenate((x,y), axis=0)
print(w)
print('----------')

# for along the axis=1 you need to have them the same in row
x=np.arange(4).reshape(2,2)
print(x)
y = np.arange(8).reshape(2,4)
print(y)
z = np.concatenate((x,y), axis=1)
print(z)
# horizonat stacking is the same a concatenate around 1 axis
print('----------')

# depth stacking
x=np.arange(4).reshape(2,2)
y = x * 2
depth_stack = np.dstack((x,y))
print(x)
print(y)
print(depth_stack) # now it has 2 in depth
print(depth_stack.shape)
print('----------')

x=np.arange(4).reshape(2,2,1)
y = x * 2
depth_stack = np.dstack((x,y))
print(x)
print(y)
print(depth_stack) # now it has 2 in depth
print(depth_stack.shape)
print('----------')

# column stacking is the same as horizontal stacking
x=np.arange(4).reshape(2,2)
print(x)
y = x * 2
print(y)
print(np.column_stack((x,y)))
print(np.column_stack((x,y)) == np.hstack((x,y)))
print('----------')


# row stacking is the same as vertical stacking
x=np.arange(4).reshape(2,2)
print(x)
y = x * 2
print(y)
print(np.row_stack((x,y)))
print(np.row_stack((x,y)) == np.vstack((x,y)))
print('----------')
```
